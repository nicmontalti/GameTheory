```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a = np.load('data/alpha.npy')

fig = plt.figure()
ax = fig.add_subplot()

ax.errorbar(a[-1]-0.01, a[2], a[3], linestyle='', marker='s', color='tab:green', label='Cooperators', capsize=3)
ax.errorbar(a[-1]+0.01, a[0], a[1], linestyle='', marker='o', color='tab:red', label='Defectors', capsize=3)
ax.legend()
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel('Fraction of players')
ax.set_xticks(a[-1])
ax.grid()

fig.savefig('plots/alpha.pdf')
```

```python
games = np.load('data/alpha_eps.npy')
Ts = np.linspace(1.1, 1.9, 10)

fig = plt.figure()
axs = fig.subplots(2,1)

colors = ['tab:orange', 'tab:olive', 'tab:cyan']
names = ['PD', 'OPD', 'PDPA']
for game, color, name in zip(games, colors, names):
        axs[0].errorbar(Ts, game[0], game[1], color=color, label=name, capsize=2, marker='.')
        axs[1].errorbar(Ts, game[2], game[3], color=color, label=name, capsize=2, marker='.')

axs[1].set_xlabel('T')
axs[1].set_ylabel(r'$\alpha$')
axs[0].set_ylabel(r'$\epsilon$')
axs[0].legend()
axs[0].grid()
axs[1].grid()
axs[0].set_xticklabels([])

fig.savefig('plots/alpha_eps.pdf')
```
